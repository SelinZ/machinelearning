```python
def abst_words (the_data, keywords):
    abst = the_data['abstract']
    abst_key = []    
    
    for i in abst:
        if i == None:
            abst_key.append(0)
            continue
        else:
            flag = 0
            for word in keywords:
                if word in i.lower():
                    flag = 1
            abst_key.append(flag)
    return pd.Series(abst_key)

keywords = ["method", "review", "randomized", "random control", "furthermore"]
print(abst_words(data, keywords))
a = data[1:2]['abstract'].to_string()
pd.set_option('display.max_colwidth', None)
print(a)
```
